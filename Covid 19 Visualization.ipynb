The aim of this project is to make readers become aware of the devastating effect of the Corona Virus Pandemic by visualizing a graph that shows the top 20 countries by the number of deaths.

We've already collected the data and added it in the top20_deathtoll.csv file (the data was collected from [Worldometer](https://www.worldometers.info/) on December 8, 2020):

```python
import pandas as pd
```

```python
top20_deathtoll = pd.read_csv('C://Users//MY PC//Desktop//DATASETS//top20_deathtoll.csv')
```

```python
top20_deathtoll
```

We will now use a bar plot to visualize the graph

```python
import matplotlib.pyplot as plt
```

```python
top20_deathtoll.plot('Country_Other', 'Total_Deaths', kind='barh', color='red')
```

We need to maximize Data Ink ratio to make the users focus on the key information

```python
fig, ax = plt.subplots()
ax.barh(top20_deathtoll['Country_Other'],
        top20_deathtoll['Total_Deaths'])
plt.show()
```

We will edit the figure size for mobile phone users.

```python
fig, ax = plt.subplots(figsize=(4.5,6))
ax.barh(top20_deathtoll['Country_Other'], top20_deathtoll['Total_Deaths'])
plt.show()
```

```python
fig,ax= plt.subplots(figsize=(4.5,6))
# to make the bar less thick, we can use the heigh=0.45
ax.barh(top20_deathtoll['Country_Other'], top20_deathtoll['Total_Deaths'], height=0.45, color='#af0b1e')
positions = ['left','right','bottom','top']
for position in positions:
    ax.spines[position].set_visible(False)
# to move the xtick labels to the top
ax.xaxis.tick_top()
# to remove the xticks
ax.tick_params(bottom=False, top=False, left=False, right=False, axis = 'x' , colors = 'grey')
# to remove the yticks
ax.tick_params(bottom=False, top=False, left=False, right=False, axis = 'y')
# to remove some of the xtick labels, ax.set_xticks()
ax.set_xticks([0, 150000, 300000])
ax.text(x=-80000, y=23.5, s= 'The Death Toll Worldwide Is 1.5M+', weight='bold', size=17)
ax.text(x=-80000, y=22.5, s= 'Top 20 countries by death toll (December 2020)', size=12)
# To add comma to 150000, 300000
ax.set_xticklabels(['0', '150,000', '300,000'])
# to left align the country names
# first,remove the county names
ax.set_yticklabels([])
country_names = top20_deathtoll['Country_Other']
for i, country in zip(range(0,21), country_names):
    ax.text(x=-80000, y=i-0.15, s=country)
# to add a vertical line, and other effects, the alpha makes it transparent

plt.show()
```

We need to add a vertical line across the graph to aid readers in interpreting the plots

```python
fig,ax= plt.subplots(figsize=(4.5,6))
# to make the bar less thick, we can use the heigh=0.45
ax.barh(top20_deathtoll['Country_Other'], top20_deathtoll['Total_Deaths'], height=0.45, color='#af0b1e')
positions = ['left','right','bottom','top']
for position in positions:
    ax.spines[position].set_visible(False)
# to move the xtick labels to the top
ax.xaxis.tick_top()
# to remove the xticks
ax.tick_params(bottom=False, top=False, left=False, right=False, axis = 'x' , colors = 'grey')
# to remove the yticks
ax.tick_params(bottom=False, top=False, left=False, right=False, axis = 'y')
# to remove some of the xtick labels, ax.set_xticks()
ax.set_xticks([0, 150000, 300000])
ax.text(x=-80000, y=23.5, s= 'The Death Toll Worldwide Is 1.5M+', weight='bold', size=17)
ax.text(x=-80000, y=22.5, s= 'Top 20 countries by death toll (December 2020)', size=12)
# To add comma to 150000, 300000
ax.set_xticklabels(['0', '150,000', '300,000'])
# to left align the country names
# first,remove the county names
ax.set_yticklabels([])
country_names = top20_deathtoll['Country_Other']
for i, country in zip(range(0,21), country_names):
    ax.text(x=-80000, y=i-0.15, s=country)
# to add a vertical line, and other effects, the alpha makes it transparent
ax.axvline(x=150000, ymin=0.045, c='grey', alpha=0.5)

plt.show()

```

From the plot above, the country with the highest number of death as at December 2020 is The United States of America with about 300, 000 deaths. Next inline is Brazil with more than 150, 000 deaths.
